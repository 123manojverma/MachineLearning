### Multi Threading and Multi Processing

- Program 
- Process 
- Threads

1. Program -> A program is a sequence of Instructions written in programming language [E.g.:- Python,C++]

- Eg:- Google Chrome-> .exe <-Program-> Browser should work

2. Process-> A process is simply an instance of a program that is being executed 
- One process cannot corrupt another process.
- I/O Requirements [Resources]
- Increased execution time to switch b/w processes.
- Eg:- Browser,
Excel->process->Id,
App->Process

3. Thread-> A thread is a unit of execution within a process.
