```python
# while loop

l = [1,2,3,4,5]
start = 0;end  = len(l)-1
while start<end:
    sum = l[start]+l[end]
    print(sum)
    start = start +1
    end = end - 1
```

```python
n=int(input())
i=1
sum=0
while i<=n:
    sum=sum+i
    i=i+1
print(f"The sum is {sum}")
```

```python
# Prime numbers b/w 1 and 100

def prime(n):
    for i in range(2,int(n**(1/2)+1)):
        if n%i==0:
            return False
    return True
for i in range(2,101):
    if prime(i):
        print(f"{i}",end=" ")
```
