# Sets



```python
# create a set
my_set={1,2,3,4,5,1,1,1,1,2,2,3,5,4,5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set=set()
print(type(my_empty_set))
```

```python
my_set=set([1,2,3,4,5,6,1,2,3])
print(my_set)
print(type(my_set))
```

```python
# To add the elements in set
my_set.add(7)
print(my_set)
```

```python
# Remove the elements from a set
my_set.remove(3)
print(my_set)
```

```python
my_set.discard(11) # it remove the element if exist otherwise do nothing
print(my_set)
```

```python
# pop method
removed_element=my_set.pop()
print(removed_element)
```

```python
print(my_set)
```

```python
# clear all the elements
my_set.clear()
print(my_set)
```

```python
# Set Membership test
my_set={1,2,3,4,5}
print(3 in my_set)
print(10 in my_set)
```

```python
# Mathematical Operation
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

# Union
union_Set=set1.union(set2)
print(union_Set)

# Intersection
intersection_set=set1.intersection(set2)
print(intersection_set)

set1.intersection_update(set2) # it update the set1 to intersection values
print(set1) 
```

```python
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

# Difference 
print(set1.difference(set2))
```

```python
set2.difference(set1)
```

```python
# Symmetric Difference
set1.symmetric_difference(set2)
```

```python
# Sets Methods
set1={1,2,3,4,5,6}
set2={4,5,6}

# is subset
print(set1.issubset(set2))

# is superset
print(set1.issuperset(set2))
```

```python
lt=[1,2,3,4,4,5]
set(lt)
```

```python
# Counting Unique words in text

text="In this tutorial we are discussing about sets"
words=text.split()

# Convert list of words to set ot get unique words

unique_words=set(words)
print(unique_words)
print(len(unique_words))
```

# Dictonaries

#### Unordered Collection of items, key must be Unique and Immutable, Store in key value pair

```python
# Creating Dictionaries
empty_dict={}
print(type(empty_dict))
```

```python
empty_dict=dict()
print(type(empty_dict))
```

```python
lang={"name":"Python","abb":"py"}
print(lang)
print(type(lang))
```

```python
# Error
student={"name":"Krish","age":45,"name":24}
print(student) # It will overwrite the value
```

```python
# Accessing Dictionary Elements
print(lang["name"])
print(lang["abb"])

# Accessing using get() method
print(lang.get('name'))
print(lang.get('abb'))
print(lang.get("used for","machine learning")) # default value if not present
```

```python
# Modifying Dictionary Elements
# Dictionary are mutable, so you can add, update or delete elements
print(lang)
```

```python
lang["name"]="Java" ## Update value for the key
print(lang)
lang["age"]=35 # added new key and value
print(lang)
```

```python
del lang['age'] # delete key and value pair
print(lang)
```

```python
# Dictionary methods

keys=lang.keys() # get all the keys
print(keys)
values=lang.values() # get all values
print(values)

items=student.items() # get all key value pairs
print(items)
```

```python
# shallow copy

lang_copy=lang
print(lang_copy)
```

```python
lang["name"]="Typescript"
print(lang_copy)
```

```python
lang_copy=lang.copy() # shallow copy
print(lang_copy)
print(lang)
```

```python
lang["name"]="Python"
print(lang_copy)
print(lang)
```

```python
# Iterating Over Dictionaries 
# You can use loops to iterate over dictionaries, keys,values or items

# Iterating over keys
for key in lang.keys():
    print(key)
```

```python
# Iterate over values
for value in lang.values():
    print(value)
```

```python
# Iterate over key value pairs
for key,value in lang.items():
    print(f"{key}:{value}")
```

```python
# Nested Dictionaries
students={
    "student1":{"name":"Krish","age":32},
    "student2":{"name":"Peter","age":35},
}
print(students)
```

```python
# Access nested dictionaries elements
print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
students.items()
```

```python
# Iterating over nested dictionaries
for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
# Dictionary Comprehension
squares={x:x**2 for x in range(5)}
print(squares)
```

```python
# Condition dictionary comprehension
events={x:x**2 for x in range(10) if x%2==0}
print(events)
```

```python
# Practical Examples

# Use a dictionary to count the frequency of elements in list

numbers=[1,2,2,3,3,3,4,4,4,4]
freq={}
for num in numbers:
    if num in freq:
        freq[num]+=1
    else:
        freq[num]=1
print(freq)
```

```python
# Merge 2 dictionareis into one

dict1={"a":1,"b":2}
dict2={"a":3,"c":4}
merged_dict={**dict1,**dict2}
print(merged_dict)
```

```python
# Manage a to do list

todo_list=["Buy Groceries","Clean the house","Pay bills"]

# Adding to task
todo_list.append("Schedule meeting")
todo_list.append("Go For a Run")

# Removing a completed task
todo_list.remove("Clean the house")

# Checking if a task is in the list
if "Pay bills" in todo_list:
    print("Don't forget to pay the utility bills")
print("To Do List remaining")
for task in todo_list:
    print(f"-{task}")
```
